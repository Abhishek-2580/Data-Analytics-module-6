```python
#Import required Library
import pandas as pd
import scipy.stats as stats
from sklearn.datasets import load_iris
```

```python
# Load Iris for T-Test
iris_data = load_iris()
iris = pd.DataFrame(iris_data.data, columns=iris_data.feature_names)
iris['species'] = iris_data.target

# Load Titanic for Chi-Square
titanic_url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
titanic = pd.read_csv(titanic_url)
display(titanic.head())
```

```python
# Separate the two groups
group1 = iris[iris['species'] == 0]['sepal length (cm)'] # Setosa
group2 = iris[iris['species'] == 1]['sepal length (cm)'] # Versicolor

# Perform Independent T-Test
t_stat, p_val_t = stats.ttest_ind(group1, group2)

print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p_val_t}")
```

### Interpretation & ConclusionHypothesis:
 $H_0$ states there is no difference in mean sepal length;
 $H_1$ states there is a significant difference.
### Result:
 The p-value is extremely small (typically $< 0.001$).
### Conclusion:
Since $p < 0.05$, we reject the Null Hypothesis. There is a statistically significant difference in sepal lengths between these two species.

```python
# Create a contingency table
contingency_table = pd.crosstab(titanic['Pclass'], titanic['Survived'])
print("Contingency Table:\n", contingency_table)

# Perform Chi-Square Test
chi2, p_val_chi, dof, expected = stats.chi2_contingency(contingency_table)

print(f"\nChi-Square Statistic: {chi2:.4f}")
print(f"P-value: {p_val_chi}")
```

### Interpretation & ConclusionHypothesis:
 $H_0$ states that Passenger Class and Survival are independent; $H_1$ states they are dependent.
### Result:
 The p-value is significantly less than $0.05$.
### Conclusion:
 We reject the Null Hypothesis. This indicates that a passenger's class (1st, 2nd, or 3rd) had a significant impact on their likelihood of survival.
